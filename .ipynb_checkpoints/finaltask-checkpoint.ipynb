# FRUIT INSPECTION

Students should develop a software system aimed at locating defects and imperfections on fruits. Each fruit is acquired 
through a NIR (Near Infra-Red) and a color camera with little parallax effect. 
The image name format is Cx_yyyyyy.bmp, where x is 0 for the NIR images and 1 for the color images, while yyyyyy is the 
incremental number of the pair. 

## FINAL TASK: KIWI INSPECTION

Images in folder “final challenge” show 5 kiwis. <br>
Students should segment the fruits and locate the defect in image “000007”. Special care should be taken to remove as “background” the dirt on the conveyor as well as the sticker in image “000006”.

#### 1. Segmentation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

First of all, we visualize the images with their respective gray-level histogram.

```python
fig, axs = plt.subplots(5, 2, figsize=(12, 12), gridspec_kw={'width_ratios': [1, 1.5]})

for i in range(5):
    path = f'./fruit-inspection-images/final challenge/C0_0000{i+6:02d}.png'
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    hist, bins = np.histogram(img.flatten(), 256, [0,256])

    axs[i, 0].imshow(img, cmap='gray'), axs[i, 0].set_title(f"C0_0000{i+6:02d}.png")
    axs[i, 1].stem(hist)
    plt.tight_layout()

plt.show()    
```

We want to isolate the fruit, so we create a binary mask.

```python
fig, axs = plt.subplots(5, 3, figsize=(13, 13), gridspec_kw={'width_ratios': [1, 1, 1]})

for i in range(5):
    path = f'./fruit-inspection-images/final challenge/C0_0000{i+6:02d}.png'
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    #Apply Otsu's algorithm
    hist, bins = np.histogram(img.flatten(), 256, [0,256])
    median = np.ceil(np.mean(img.flatten()))
    g_blur = cv2.GaussianBlur(img,(9,9),0)
    ret,th = cv2.threshold(g_blur,median,255,cv2.THRESH_BINARY)

    #Apply adaptive thresholds
    blur = cv2.bilateralFilter(img,5,15,15)
    th1 = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,5,3)

    #fill background
    th = np.pad(th, 1, mode='constant', constant_values=255)
    cv2.floodFill(th, None, (0, 0), 0)
    th = th[1:-1, 1:-1]
    
    th1 = np.pad(th1, 1, mode='constant', constant_values=255)
    cv2.floodFill(th1, None, (0, 0), 0)
    th1 = th1[1:-1, 1:-1]

    #Flood-fill the background with 255 (white)
    th_copy = th.copy()
    h, w = img.shape
    mask = np.zeros((h+2, w+2), dtype=np.uint8)
    cv2.floodFill(th_copy, mask, (0,0), 255)
    
    th1_copy = th1.copy()
    h, w = img.shape
    mask = np.zeros((h+2, w+2), dtype=np.uint8)
    cv2.floodFill(th1_copy, mask, (0,0), 255)

    #Fill internal holes by overlapping results
    th_final = np.maximum(th, cv2.bitwise_not(th_copy))
    th1_final = np.maximum(th1, cv2.bitwise_not(th1_copy))

    fruit_only = cv2.bitwise_and(img, img, mask=th_final)
    fruit_only1 = cv2.bitwise_and(img, img, mask=th1_final)
    
    #Print images
    axs[i, 0].imshow(img, cmap='gray'), axs[i, 0].set_title(f"C0_0000{i+6:02d}.png")
    axs[i, 1].imshow(fruit_only, cmap='gray'), axs[i, 1].set_title(f"Otsu on C0_0000{i+6:02d}.png")
    axs[i, 2].imshow(fruit_only1, cmap='gray'), axs[i, 2].set_title(f"Ad th mean on C0_0000{i+6:02d}.png")

    plt.tight_layout()

plt.show()    
```

We can observe that Adaptive thresholding cannot isolate kiwis, due to their fuzzy skin which introduces micro-variations in pixel intensity that thresholding might misinterpret as edges or background. But it can isolate the sticker in image "000006". Otsu's algorithm instead can isolate the fruit from background, but also the sticker. By subtracting the mask generated by adaptive thresholding from the mask produced by Otsu's algorithm, we can effectively isolate the fruit.

```python
fig, axs = plt.subplots(5, 2, figsize=(13, 13), gridspec_kw={'width_ratios': [1, 1]})

th_final_arr = []

for i in range(5):
    path = f'./fruit-inspection-images/final challenge/C0_0000{i+6:02d}.png'
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    hist, bins = np.histogram(img.flatten(), 256, [0,256])
    median = np.ceil(np.mean(img.flatten()))
    #print(median)

    #Apply Otsu's algorithm
    g_blur = cv2.GaussianBlur(img,(9,9),0)
    #ret,th = cv2.threshold(g_blur,35,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    ret,th = cv2.threshold(g_blur,median,255,cv2.THRESH_BINARY) #THIS ONE WORKS ALMOST FINE

    #Apply adaptive thresholds
    blur = cv2.bilateralFilter(img,5,15,15)
    th1 = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,5,3)

    #fill background
    th = np.pad(th, 1, mode='constant', constant_values=255)
    cv2.floodFill(th, None, (0, 0), 0)
    th = th[1:-1, 1:-1]
    
    th1 = np.pad(th1, 1, mode='constant', constant_values=255)
    cv2.floodFill(th1, None, (0, 0), 0)
    th1 = th1[1:-1, 1:-1]

    #Flood-fill the background with 255 (white)
    th_copy = th.copy()
    h, w = img.shape
    mask = np.zeros((h+2, w+2), dtype=np.uint8)
    cv2.floodFill(th_copy, mask, (0,0), 255)
    
    th1_copy = th1.copy()
    h, w = img.shape
    mask = np.zeros((h+2, w+2), dtype=np.uint8)
    cv2.floodFill(th1_copy, mask, (0,0), 255)

    #Fill internal holes by overlapping results
    th_final = np.maximum(th, cv2.bitwise_not(th_copy))
    th1_final = np.maximum(th1, cv2.bitwise_not(th1_copy))

    # Subtract Adaptive Thresh generated mask from Otsu generated mask
    th_final = cv2.bitwise_and(th, th, mask=cv2.bitwise_not(th1))    
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(th_final, connectivity=8)
    
    # Search largest area that corrispond to the kiwi
    max_area = 0
    largest_label = 0
    
    for label in range(1, num_labels):
        area = stats[label, cv2.CC_STAT_AREA]
        if area > max_area:
            max_area = area
            largest_label = label
    
    # Create a new image
    output_image = np.zeros_like(th_final, dtype=np.uint8)
    
    # Take the largest area
    output_image[labels == largest_label] = 255

    element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    output_image = cv2.morphologyEx(output_image, cv2.MORPH_CLOSE, element)

    ##############################################

    fruit_only = cv2.bitwise_and(img, img, mask=output_image)

    th_final_arr.append(output_image)
    
    #Print images
    axs[i, 0].imshow(img, cmap='gray'), axs[i, 0].set_title(f"C0_0000{i+6:02d}.png")
    axs[i, 1].imshow(fruit_only, cmap='gray'), axs[i, 1].set_title(f"Only fruit C0_0000{i+6:02d}.png")

    plt.tight_layout()

plt.show()    
```

We can apply the mask to the colored image because of the parallax between NIR and colored image.

```python
fig, axs = plt.subplots(5, 2, figsize=(13, 13), gridspec_kw={'width_ratios': [1, 1]})

fruit_only_arr = []

for i in range(5):
    path = f'./fruit-inspection-images/final challenge/C1_0000{i+6:02d}.png'
    img = cv2.imread(path)

    fruit_only = cv2.bitwise_and(img, img, mask=th_final_arr[i])

    fruit_only_arr.append(fruit_only)
    
    #Print images
    axs[i, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), axs[i, 0].set_title(f"C0_0000{i+6:02d}.png")
    axs[i, 1].imshow(cv2.cvtColor(fruit_only, cv2.COLOR_BGR2RGB)), axs[i, 1].set_title(f"Only fruit C0_0000{i+6:02d}.png")

    plt.tight_layout()

plt.show()    
```

#### 2. Locate defects.

For defects detection we will use the same method used in the first task. We will use Canny's edge detector twice, the first time we will find the external edge, then we will use two different thresholds that allow us to find all the edges. Finally we will subtract to them the external edge found earlier in order to obtain the internal edges.

```python
fig, axs = plt.subplots(5, 5, figsize=(13, 13), gridspec_kw={'width_ratios': [1, 1, 1, 1, 1]})

int_edge_dil_arr = []
for i in range(5):
    path = f'./fruit-inspection-images/final challenge/C0_0000{i+6:02d}.png'
    img = cv2.imread(path)
    g_blur = cv2.GaussianBlur(img,(9,9),0)
    #fruit_only1 = cv2.bitwise_and(g_blur, g_blur, mask=filled_th1_arr[0])
    mat = np.ones_like(img)*255
    fruit_only1 = cv2.bitwise_not(cv2.bitwise_and(mat, cv2.bitwise_not(g_blur), mask=th_final_arr[i]))
    fruit_only1 = cv2.GaussianBlur(fruit_only1,(7,7),0)
    ext_edge = cv2.Canny(fruit_only1, 200, 300)
    kernel = np.ones((10,10))
    ext_edge_dil = cv2.dilate(ext_edge, kernel, iterations=1)
    
    all_edge = cv2.Canny(fruit_only1, 40, 80)
    int_edge = cv2.bitwise_and(all_edge, all_edge, mask=cv2.bitwise_not(ext_edge_dil))

    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (50, 50))
    int_edge_dil = cv2.morphologyEx(int_edge, cv2.MORPH_CLOSE, kernel)
    kernel = np.ones((3,3))
    int_edge_dil = cv2.erode(int_edge_dil, kernel, iterations=1)
    int_edge_dil_arr.append(int_edge_dil)

    # Create a red image
    edge_color=(255, 0, 0)
    alpha = 0.5
    colored_edges = np.zeros_like(img)
    colored_edges[int_edge_dil == 255] = edge_color
    
    # Apply overlapping
    defects = img.copy()
    mask_indices = int_edge_dil == 255  # Trova i pixel dove applicare l'overlay
    for c in range(3):  # Applica per ciascun canale (R, G, B)
        defects[..., c][mask_indices] = (
            (1 - alpha) * img[..., c][mask_indices] + alpha * colored_edges[..., c][mask_indices]
        ).astype(np.uint8)

    #Print images
    axs[i, 0].imshow(img), axs[i, 0].set_title(f"C0_00000{i+1}.png")
    axs[i, 1].imshow(all_edge, cmap='gray'), axs[i, 1].set_title(f"All edges C0_00000{i+1}.png")
    axs[i, 2].imshow(ext_edge, cmap='gray'), axs[i, 2].set_title(f"External edges C0_00000{i+1}.png")
    axs[i, 3].imshow(int_edge_dil, cmap='gray'), axs[i, 3].set_title(f"Internal edges C0_00000{i+1}.png")
    axs[i, 4].imshow(defects, cmap='gray'), axs[i, 4].set_title(f"Defects of C0_00000{i+1}.png")

    plt.tight_layout()
```

```python
fig, axs = plt.subplots(5, 2, figsize=(15, 15), gridspec_kw={'width_ratios': [1, 1]})
for i in range(5):
    defects = fruit_only_arr[i].copy()

    # Draw a circle on defect
    contours, _ = cv2.findContours(int_edge_dil_arr[i], cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        (center_x, center_y), radius = cv2.minEnclosingCircle(contour)
        center = (int(center_x), int(center_y))
        radius = int(radius)
        # Draw on image
        cv2.circle(defects, center, radius, (0, 255, 0), 2)

    #Print images
    axs[i, 0].imshow(cv2.cvtColor(fruit_only_arr[i], cv2.COLOR_BGR2RGB)), axs[i, 0].set_title(f"C1_00000{i+1}.png")
    axs[i, 1].imshow(cv2.cvtColor(defects, cv2.COLOR_BGR2RGB)), axs[i, 1].set_title(f"Defects of C1_00000{i+1}.png")

    plt.tight_layout()
```
